```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
data=pd.read_csv("countries.csv")
```

```python
data.head(10)
```

```python
data_1997=data[data.year==1997]
```

```python
data_1997.tail(10)
```

```python
set(data.continent)
```

```python
Americas_1997=data_1997[data_1997.continent=="Americas"]
```

```python
Europe_1997=data_1997[data_1997.continent=="Europe"]
```

```python
plt.subplot(2,1,1)
plt.hist(Americas_1997.lifeExpectancy,30,range=(50,90),edgecolor="black")
plt.ylabel("America")
plt.subplot(2,1,2)
plt.hist(Europe_1997.lifeExpectancy,30,range=(50,90),edgecolor="black")
plt.ylabel("Europe")
plt.show()
```
