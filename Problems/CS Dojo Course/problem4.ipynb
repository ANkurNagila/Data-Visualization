```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("countries.csv")
```

```python
data.head(10)
```

```python
data_2007=data[data.year==2007]
```

```python
top_10=data_2007.sort_values("population",ascending=False).head(10)
```

```python
top_10
```

```python
GDP=top_10.gdpPerCapita*top_10.population/10**12
```

```python
GDP
```

```python
plt.subplot(2,1,1)
x=range(10)
plt.bar(x,top_10.population/10**9)
plt.title("Comparing GDP of Top 10 populous countries of 2007")
plt.xticks([],[])
plt.ylabel("Population in Millions")
plt.subplot(2,1,2)
x=range(10)
plt.bar(x,GDP)
plt.xticks(x,top_10.country,rotation="vertical")
plt.ylabel("GDP in Billions")
plt.show()
```
